```python
import os
import numpy as np
from oads_access.oads_access import OADS_Access
from PIL import Image
from skimage.transform import warp_polar
from matplotlib import pyplot as plt
```

```python
oads = OADS_Access(basedir='/home/niklas/projects/data/oads/')
```

```python
rng = np.random.default_rng()
```

```python
image_name = rng.choice(list(oads.image_names.keys()))
```

```python
img, label = oads.load_image(image_name=image_name)
```

```python
img = np.array(img.reduce(8).resize((450,450)))
```

```python
img.shape
```

```python
center = (225, 225)
```

```python
warp = warp_polar(img, channel_axis=2, center=center, scaling='log', radius=225)
```

```python
warp.shape
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(img)
ax[1].imshow(warp)
plt.show()
```
